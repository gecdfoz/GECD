#### Quais são as principais causas dos acidentes com vítimas fatais?

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df[df['mortos'] > 0]['causa_acidente'].value_counts()
```

```python
def count_plot_causa_acidente(data):
    chart = sns.countplot(
        y='causa_acidente',
        data=data,
        orient='v',
        order=data['causa_acidente'].value_counts().index,
    )
    chart.axes.set_ylabel('Causa dos acidentes')
    plt.show()
count_plot_causa_acidente(df[df['mortos'] > 0])
```

#### E no Paraná?

```python
count_plot_causa_acidente(acidentes_parana[acidentes_parana['mortos'] > 0])
```

#### E em Foz?

```python
count_plot_causa_acidente(acidentes_parana[(acidentes_parana['mortos'] > 0) & (acidentes_parana['municipio'] == 'FOZ DO IGUACU')])
```

```python
data = df[(df['municipio'].isin(['FOZ DO IGUACU', 'GUARULHOS']))].groupby(['municipio'])['causa_acidente'].value_counts(normalize=True).rename('percentage').mul(100).reset_index()
data
```

```python
chart = sns.barplot(
    y='causa_acidente',
    x='percentage',
    data=data,
    hue='municipio',
)
chart.axes.set_ylabel('Causa dos acidentes')
plt.show()
```

```python
data = data[data['causa_acidente'].isin(
    ['Falta de Atenção à Condução', 'Animais na Pista',
     'Não guardar distância de segurança',
       'Desobediência às normas de trânsito pelo condutor',
       'Ingestão de Álcool', 'Velocidade Incompatível',
       'Pista Escorregadia', 'Condutor Dormindo',
       'Defeito Mecânico no Veículo',
       'Restrição de Visibilidade',
       'Falta de Atenção do Pedestre',
       'Mal Súbito', 'Agressão Externa',
    ]
)]

chart = sns.barplot(
    y='causa_acidente',
    x='percentage',
    data=data,
    hue='municipio',
)

sns.despine(top=False, right=False)
sns.set_style("whitegrid")

chart.axes.set_ylabel('Causa dos acidentes', fontsize=30)
chart.axes.set_xlabel('Número de acidentes (%)', fontsize=30)
chart.tick_params(labelsize=18)
chart.grid(False)

sns.set(rc={'figure.figsize':(20,15)})

plt.legend(loc='center right')
plt.setp(chart.get_legend().get_texts(), fontsize='22')



plt.show()
```

## Mapa de calor

A biblioteca Folium é feita para visualizações usando mapas

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
```

#### Conversão de latitude e longitude para número

Isso vai acontecer em várias bases! Como os Estados Unidos usa padrão diferente da maioria dos países (com ponto ao invés de vírgula) o Pandas não reconhece o tipo, então temos que converter substituir e converter

```python
df['latitude'] = pd.to_numeric(df['latitude'].str.replace(",", "."))
df['longitude'] = pd.to_numeric(df['longitude'].str.replace(",", "."))
```

```python
mapa = folium.Map(
    location=[-24.5952516, -53.5713044],
    zoom_start=7
)
```

```python
mapa
```

#### Remoção dos valores nulos

```python
df=df.dropna(subset=['latitude'])
df=df.dropna(subset=['longitude'])
```

```python
acidentes_parana = df[df['uf'] == 'PR']
acidentes_parana[['latitude', 'longitude']].count()
```

```python
coordenadas_dos_acidentes = acidentes_parana[['latitude', 'longitude']].values
HeatMap(coordenadas_dos_acidentes).add_to(mapa)
mapa
```
